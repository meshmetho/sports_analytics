```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#plot 1
x = np.array([1903, 1919, 1925, 1934, 1951, 1961, 1974, 1980, 1994, 2005, 2015])
y = np.array([94.55, 231.12, 219.17, 147.23, 142.23, 122.03, 116.28, 109.32, 103.64, 86.25, 84.77])
plt.subplot(2,1,1)
plt.plot(x, y)
plt.xlabel("Years")
plt.ylabel("Time Overall")

#plot 2
x = np.array([1903, 1919, 1925, 1934, 1951, 1961, 1974, 1980, 1994, 2005, 2015])
y = np.array([2.99, 1.71, 0.92, 0.46, 0.37, 0.2, 0.14, 0.1, 0.08, 0.07, 0.02])
plt.subplot(2,1,2)
plt.xlabel("Years")
plt.ylabel("Time Margin")
plt.plot(x, y)
plt.tight_layout(pad = 3.0)
plt.show()
```

From the data given we can see that both time taken and time margin between the winner and runners up decreased over the years. This is mainly due research and basic structural design changes done over the years. Such structural changes include fabrication of materials(using lighter materials eg aluminium and its alloys), introduction of shock absorbers and gearing mechanisms.  These changes have led to better bicycles being manufactured hence increasing their speed and making them more withstanding to the various terrains on the racing tracks

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
x = np.array(["France", "Luxembourg", "Belgium", "Italy", "Switzerland", "Colombia", "Spain", "Netherlands", "USA", "Ireland", "Denmark", "Germany", "Great Britain", "Australia"])
y = np.array([36, 5, 18, 10, 2, 1, 12, 2, 10, 1, 1, 1, 6, 1])

plt.title("Nationalities")
plt.xlabel("No. of times won")
plt.barh(x, y)
plt.show
```

From the data above we can clearly see that throughout the years France has produced most winners in the competition. This might be due to the competition being primarily held in France hence the racers do have home advantage
